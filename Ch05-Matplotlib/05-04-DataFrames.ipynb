```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from matplotlib import pyplot as plt
ts = Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
#display(ts)
plt.plot(ts)
plt.show()
ts.plot()
plt.show()
#plt.gcf().canvas.get_supported_filetypes_grouped()
```

```python
ts = Series(np.random.randn(1000), \
            index=pd.date_range('1/1/2000', periods=1000))
df = DataFrame(np.random.randn(1000,4), \
index= ts.index, columns='AAPL,MSFT,GOOG,HP'.split(','))
df = df.cumsum()
#display(df)
#df.plot(color = ['red', 'blue', 'green'])
#plt.plot(df)

#display(df[['AAPL', 'MSFT']])
df[['AAPL', 'MSFT']].plot(kind='line')
```
